```python
# 1. Leer el archivo CSV masivo desde nuestro Data Lake
df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("Files/Retail_Transactions_Dataset.csv")

# 2. Mostrar las primeras filas en pantalla para verificar
display(df)
```

```python
from pyspark.sql.functions import col, year, month, to_date

# 1. Transformación (T): Asegurar formato de fecha y extraer Año y Mes
# Convertimos la columna Date a formato Fecha real de base de datos
df_transformado = df.withColumn("Date", to_date(col("Date"))) \
                    .withColumn("Year", year(col("Date"))) \
                    .withColumn("Month", month(col("Date")))

# 2. Carga (L): Guardar como Tabla Delta en el Lakehouse
# El formato "delta" crea una tabla SQL super rápida y optimizada
df_transformado.write.format("delta").mode("overwrite").saveAsTable("ventas_retail_procesadas")

print("¡Proceso ETL completado con éxito! El CSV ahora es una Tabla Delta lista para SQL.")
```

```python
%%sql
-- Usamos SQL puro para analizar los datos que acabamos de limpiar con Python
SELECT 
    City, 
    COUNT(Transaction_ID) as Cantidad_Ventas
FROM ventas_retail_procesadas
GROUP BY City
ORDER BY Cantidad_Ventas DESC
LIMIT 10;
```
